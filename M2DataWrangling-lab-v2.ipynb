### 4. Encoding Categorical Variables


<h5>4.1 Encode the Employment column using one-hot encoding.</h5>


```python
df_encoded = pd.get_dummies(df, columns=['Employment'], prefix='Employment')

# Display the resulting DataFrame
print(df_encoded)
```

### 5. Handling Missing Values


<h5>5.1 Identify columns with the highest number of missing values.</h5>


```python
# Identify columns with the highest number of missing values
missing_values = df.isnull().sum()

# Sort columns by the number of missing values
missing_values_sorted = missing_values.sort_values(ascending=False)

# Display the columns with the highest missing values
print(missing_values_sorted)
```

<h5>5.2 Impute missing values in numerical columns (e.g., `ConvertedCompYearly`) with the mean or median.</h5>


```python
df['ConvertedCompYearly'] = df['ConvertedCompYearly'].fillna(df['ConvertedCompYearly'].mean())

# Option 2: Alternatively, you can use median to fill missing values
# df['ConvertedCompYearly'] = df['ConvertedCompYearly'].fillna(df['ConvertedCompYearly'].median())

# Display the DataFrame after imputation
print(df)
```

<h5>5.3 Impute missing values in categorical columns (e.g., `RemoteWork`) with the most frequent value.</h5>


```python
# Impute missing values in categorical columns using the most frequent value
df['RemoteWork'] = df['RemoteWork'].fillna(df['RemoteWork'].mode()[0])

# Display the DataFrame after imputation
print(df)
```

### 6. Feature Scaling and Transformation


<h5>6.1 Apply Min-Max Scaling to normalize the `ConvertedCompYearly` column.</h5>


```python
from sklearn.preprocessing import MinMaxScaler

# Initialize MinMaxScaler
scaler = MinMaxScaler()

# Apply Min-Max Scaling to normalize the 'ConvertedCompYearly' column
df['ConvertedCompYearly'] = scaler.fit_transform(df[['ConvertedCompYearly']])

# Display the DataFrame after scaling
print(df)
```

<h5>6.2 Log-transform the ConvertedCompYearly column to reduce skewness.</h5>


```python
import numpy as np

# Apply log transformation to the 'ConvertedCompYearly' column
df['ConvertedCompYearly'] = np.log1p(df['ConvertedCompYearly'])

# Display the DataFrame after log transformation
print(df)
```

### 7. Feature Engineering


<h5>7.1 Create a new column `ExperienceLevel` based on the `YearsCodePro` column:</h5>


```python
# Ensure the 'YearsCodePro' column is numeric, forcing errors to NaN if conversion fails
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')

# Define the conditions for the ExperienceLevel column
conditions = [
    (df['YearsCodePro'] < 1),  # Less than 1 year: Junior
    (df['YearsCodePro'] >= 1) & (df['YearsCodePro'] < 3),  # 1 to 3 years: Mid-level
    (df['YearsCodePro'] >= 3) & (df['YearsCodePro'] < 6),  # 3 to 6 years: Senior
    (df['YearsCodePro'] >= 6)  # 6 or more years: Expert
]

# Define the corresponding labels for each condition
labels = ['Junior', 'Mid-level', 'Senior', 'Expert']

# Create the new 'ExperienceLevel' column based on the conditions
df['ExperienceLevel'] = pd.cut(df['YearsCodePro'], bins=[-float('inf'), 1, 3, 6, float('inf')], labels=labels)

# Display the DataFrame with the new column
print(df)
```
