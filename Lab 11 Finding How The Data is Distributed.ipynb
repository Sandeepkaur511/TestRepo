### Step 1: Import Libraries and Load Data

### Step 3: Handle Missing Data


- Identify missing values in the dataset.

- Impute or remove missing values as necessary to ensure data completeness.



```python
# Step 1: Identify missing values in all columns
missing_values = df.isnull().sum()
print("Missing values per column:")
print(missing_values)

# Impute missing values in 'EdLevel' with the most frequent value (mode)
most_frequent_value = df['EdLevel'].mode()[0]
df['EdLevel'].fillna(most_frequent_value, inplace=True)

# Step 3: Verify that there are no missing values in 'EdLevel' after imputation
missing_values_after_imputation = df['EdLevel'].isnull().sum()
print(f"Missing values in 'EdLevel' after imputation: {missing_values_after_imputation}")
# Step 1: Check for missing values in 'ConvertedCompYearly' column
missing_compensation = df['ConvertedCompYearly'].isnull().sum()
print(f"Missing values in 'ConvertedCompYearly': {missing_compensation}")

# Step 2: If missing values exist, handle them (e.g., imputing with the median compensation)
if missing_compensation > 0:
    median_compensation = df['ConvertedCompYearly'].median()
    df['ConvertedCompYearly'].fillna(median_compensation, inplace=True)
    print(f"Missing values in 'ConvertedCompYearly' have been imputed with the median value: {median_compensation}")

# Step 3: Verify that there are no missing values in 'ConvertedCompYearly' after imputation
missing_compensation_after = df['ConvertedCompYearly'].isnull().sum()
print(f"Missing values in 'ConvertedCompYearly' after imputation: {missing_compensation_after}")
```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
!pip install matplotlib-venn
import matplotlib.pyplot as plt
from collections import Counter
from matplotlib_venn import venn2

# Step 6: Compare LanguageHaveWorkedWith vs LanguageWantToWorkWith

# Drop missing values and split the strings into individual languages
worked_langs = df['LanguageHaveWorkedWith'].dropna().str.split(';')
want_langs = df['LanguageWantToWorkWith'].dropna().str.split(';')

# Flatten the lists
worked_flat = [lang.strip() for sublist in worked_langs for lang in sublist]
want_flat = [lang.strip() for sublist in want_langs for lang in sublist]

# Count frequencies
worked_counts = Counter(worked_flat)
want_counts = Counter(want_flat)

# Venn diagram of unique languages
plt.figure(figsize=(8, 6))
venn2([set(worked_flat), set(want_flat)], set_labels=('Have Worked With', 'Want To Work With'))
plt.title('Venn Diagram of Programming Languages')
plt.show()

# Optional: Grouped Bar Chart of top common languages
top_languages = set(worked_counts.keys()).intersection(set(want_counts.keys()))
top_languages = sorted(top_languages, key=lambda x: worked_counts[x] + want_counts[x], reverse=True)[:10]

worked_vals = [worked_counts[lang] for lang in top_languages]
want_vals = [want_counts[lang] for lang in top_languages]

x = range(len(top_languages))
plt.figure(figsize=(12, 6))
plt.bar(x, worked_vals, width=0.4, label='Have Worked With', align='center')
plt.bar([i + 0.4 for i in x], want_vals, width=0.4, label='Want To Work With', align='center')
plt.xticks([i + 0.2 for i in x], top_languages, rotation=45)
plt.title('Top Programming Languages: Worked With vs Want To Work With')
plt.legend()
plt.tight_layout()
plt.show()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
import pandas as pd
from scipy.stats import pearsonr, spearmanr

# Ensure that the columns are numeric
df['JobSat'] = pd.to_numeric(df['JobSat'], errors='coerce')
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')

# Drop rows with NaN values in the 'JobSat' and 'YearsCodePro' columns
df_cleaned = df.dropna(subset=['JobSat', 'YearsCodePro'])

# Calculate Pearson Correlation
pearson_corr, _ = pearsonr(df_cleaned['JobSat'], df_cleaned['YearsCodePro'])
print(f"Pearson Correlation Coefficient: {pearson_corr:.4f}")

# Calculate Spearman Correlation
spearman_corr, _ = spearmanr(df_cleaned['JobSat'], df_cleaned['YearsCodePro'])
print(f"Spearman Correlation Coefficient: {spearman_corr:.4f}")

```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
# import pandas as pd
# import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame
# Create a cross-tabulation between 'Employment' and 'EdLevel'
cross_tab = pd.crosstab(df['Employment'], df['EdLevel'])

# Display the cross-tabulation
print(cross_tab)

# Plotting the cross-tabulation as a stacked bar plot
cross_tab.plot(kind='bar', stacked=True, figsize=(10, 6), color=plt.cm.Paired.colors)

# Adding labels and title
plt.title('Employment Status vs Education Level')
plt.xlabel('Employment Status')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Education Level', bbox_to_anchor=(1, 1), loc='upper left')

# Show the plot
plt.tight_layout()
plt.show()
```
