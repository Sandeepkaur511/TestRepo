## Hands on Lab


##### Step 1: Install and Import Required Libraries


```python
# Install the necessary libraries
!pip install pandas
!pip install matplotlib
!pip install seaborn

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2: Load the Dataset


```python
# Load the dataset from the given URL
file_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_url)

# Display the first few rows to understand the structure of the dataset
df.head()
```

<h3>Step 3: Analyze and Visualize Compensation Distribution</h3>


**Task**: Plot the distribution and histogram for `ConvertedCompYearly` to examine the spread of yearly compensation among respondents.


```python
# Drop NaN values first to avoid issues
comp_data = df['ConvertedCompYearly'].dropna()

# Set up the figure
plt.figure(figsize=(12, 6))

# Plot the distribution with a histogram + KDE curve
sns.histplot(comp_data, kde=True, color='skyblue', bins=30)

# Add labels and title
plt.title('Distribution of Converted Compensation Yearly')
plt.xlabel('Yearly Compensation (USD)')
plt.ylabel('Number of Respondents')

# Show plot
plt.show()
```

<h3>Step 4: Calculate Median Compensation for Full-Time Employees</h3>


**Task**: Filter the data to calculate the median compensation for respondents whose employment status is "Employed, full-time."


```python
# Filter data for 'Employed, full-time'
full_time_df = df[df['Employment'] == 'Employed, full-time']

# Calculate the median compensation
median_comp_full_time = full_time_df['ConvertedCompYearly'].median()

# Print the result
print(f"Median Yearly Compensation for Full-Time Employees: ${median_comp_full_time:,.2f}")
```

<h3>Step 5: Analyzing Compensation Range and Distribution by Country</h3>


Explore the range of compensation in the ConvertedCompYearly column by analyzing differences across countries. Use box plots to compare the compensation distributions for each country to identify variations and anomalies within each region, providing insights into global compensation trends.



```python
 # Drop rows with missing compensation or country
country_comp_df = df.dropna(subset=['ConvertedCompYearly', 'Country'])

# Set up the figure size
plt.figure(figsize=(16, 8))

# Create a box plot grouped by Country
sns.boxplot(data=country_comp_df, x='ConvertedCompYearly', y='Country', palette='coolwarm')

# Add labels and title
plt.title('Yearly Compensation Distribution by Country')
plt.xlabel('Converted Compensation Yearly (USD)')
plt.ylabel('Country')

# Optional: set x-axis to log scale if the data is very skewed
plt.xscale('log')  # Comment this out if you prefer linear

# Show the plot
plt.tight_layout()
plt.show()
```

<h3>Step 6: Removing Outliers from the Dataset</h3>


**Task**: Create a new DataFrame by removing outliers from the `ConvertedCompYearly` column to get a refined dataset for correlation analysis.


```python
# Step 1: Calculate Q1, Q3, and IQR
Q1 = df['ConvertedCompYearly'].quantile(0.25)
Q3 = df['ConvertedCompYearly'].quantile(0.75)
IQR = Q3 - Q1

# Step 2: Determine the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Step 3: Create a new DataFrame excluding outliers
refined_df = df[(df['ConvertedCompYearly'] >= lower_bound) & (df['ConvertedCompYearly'] <= upper_bound)]

# Step 4: Validate the size of the new refined DataFrame
print(f"Original dataset size: {df.shape[0]}")
print(f"Refined dataset size (without outliers): {refined_df.shape[0]}")
```

<h3>Step 7: Finding Correlations Between Key Variables</h3>


**Task**: Calculate correlations between `ConvertedCompYearly`, `WorkExp`, and `JobSatPoints_1`. Visualize these correlations with a heatmap.


```python
# Select relevant columns for correlation
corr_columns = ['ConvertedCompYearly', 'WorkExp', 'JobSatPoints_1']

# Drop rows with NaN values in these columns
corr_df = refined_df[corr_columns].dropna()

# Compute the correlation matrix
corr_matrix = corr_df.corr()

# Display the correlation matrix
print(corr_matrix)

# Visualize the correlation matrix as a heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title('Correlation Heatmap: Compensation, Work Experience, and Job Satisfaction')
plt.show()
```

<h3>Step 8: Scatter Plot for Correlations</h3>


**Task**: Create scatter plots to examine specific correlations between `ConvertedCompYearly` and `WorkExp`, as well as between `ConvertedCompYearly` and `JobSatPoints_1`.


```python
# Drop NaNs for relevant columns
scatter_df = refined_df[['ConvertedCompYearly', 'WorkExp', 'JobSatPoints_1']].dropna()

# Set up the figure with two subplots side by side
plt.figure(figsize=(14, 6))

# Scatter plot: Compensation vs Work Experience
plt.subplot(1, 2, 1)
sns.scatterplot(data=scatter_df, x='WorkExp', y='ConvertedCompYearly', color='teal')
plt.title('Compensation vs Work Experience')
plt.xlabel('Work Experience (Years)')
plt.ylabel('Converted Compensation Yearly (USD)')
plt.yscale('log')  # Optional: use log scale if compensation is skewed

# Scatter plot: Compensation vs Job Satisfaction Points
plt.subplot(1, 2, 2)
sns.scatterplot(data=scatter_df, x='JobSatPoints_1', y='ConvertedCompYearly', color='purple')
plt.title('Compensation vs Job Satisfaction')
plt.xlabel('Job Satisfaction Score')
plt.ylabel('Converted Compensation Yearly (USD)')
plt.yscale('log')  # Optional: log scale here too

# Adjust layout and display
plt.tight_layout()
plt.show()
```
