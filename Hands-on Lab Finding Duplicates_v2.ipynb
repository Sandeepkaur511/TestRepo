## Identify and Analyze Duplicates


### Task 1: Identify Duplicate Rows
1. Count the number of duplicate rows in the dataset.
3. Display the first few duplicate rows to understand their structure.


```python
## Write your code here
# Count the number of duplicate rows
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")

# Display the first few duplicate rows
duplicates = df[df.duplicated()]
duplicates.head()
```

### Task 2: Analyze Characteristics of Duplicates
1. Identify duplicate rows based on selected columns such as MainBranch, Employment, and RemoteWork. Analyse which columns frequently contain identical values within these duplicate rows.
2. Analyse the characteristics of rows that are duplicates based on a subset of columns, such as MainBranch, Employment, and RemoteWork. Determine which columns frequently have identical values across these rows.
   


```python
## Write your code here
# Step 1: Identify duplicate rows based on specific columns
subset_columns = ['MainBranch', 'Employment', 'RemoteWork']
duplicate_subset = df[df.duplicated(subset=subset_columns, keep=False)]

# Step 2: Display a few of these rows to inspect
print("Sample of duplicate rows based on subset:")
display(duplicate_subset.head())

# Step 3: Group and analyze identical value frequency in other columns
# Check how many times each column has the same value within each duplicate group
from collections import Counter

# Count of columns having same values across duplicate groups
identical_columns = []

# Group by the subset columns
grouped = duplicate_subset.groupby(subset_columns)

for _, group in grouped:
    if len(group) > 1:
        # For each column, check if all values in the group are the same
        for col in df.columns:
            if col not in subset_columns:
                if group[col].nunique() == 1:
                    identical_columns.append(col)

# Frequency of identical columns
identical_columns_count = Counter(identical_columns)

# Display the most commonly identical columns among duplicate groups
print("Columns that frequently have identical values in duplicate groups:")
for col, count in identical_columns_count.most_common():
    print(f"{col}: {count} times")
```

### Task 3: Visualize Duplicates Distribution
1. Create visualizations to show the distribution of duplicates across different categories.
2. Use bar charts or pie charts to represent the distribution of duplicates by Country and Employment.


```python
!pip install seaborn
```

```python
## Write your code here
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Get all duplicate rows (based on full row)
duplicates_all = df[df.duplicated(keep=False)]

# Step 2: Plot distribution by Country (Bar Chart)
plt.figure(figsize=(12, 6))
country_counts = duplicates_all['Country'].value_counts().head(10)  # top 10 for clarity
sns.barplot(x=country_counts.values, y=country_counts.index, palette="viridis")
plt.title("Top 10 Countries with Duplicate Entries")
plt.xlabel("Number of Duplicate Rows")
plt.ylabel("Country")
plt.tight_layout()
plt.show()

# Step 3: Plot distribution by Employment (Pie Chart)
employment_counts = duplicates_all['Employment'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(employment_counts, labels=employment_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title("Duplicate Rows by Employment Type")
plt.tight_layout()
plt.show()
```

### Task 4: Strategic Removal of Duplicates
1. Decide which columns are critical for defining uniqueness in the dataset.
2. Remove duplicates based on a subset of columns if complete row duplication is not a good criterion.


```python
## Write your code here
df.columns
# Define subset of columns that reflect core respondent identity
key_columns = ['MainBranch', 'Employment', 'RemoteWork', 'Country', 'EdLevel', 'YearsCode', 'DevType']

# Check number of duplicates based on these key columns
subset_duplicate_count = df.duplicated(subset=key_columns).sum()
print(f"Number of duplicates based on key columns: {subset_duplicate_count}")

# Remove duplicates based on key columns
df_unique = df.drop_duplicates(subset=key_columns, keep='first')

# Final shape after removing partial duplicates
print(f"Shape after removing subset duplicates: {df_unique.shape}")
print("Original shape:", df.shape)
print("New shape:", df_unique.shape)

```
